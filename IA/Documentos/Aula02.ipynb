```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
km = pd.Series ([4,5,5,5,5,5,5,6,8,8,9,10,10,12,12,12,17,1,18,19,20,21,21,21,21,25,26,27,28,30,31,33,34,36,42,42])
```

```python
km.shape, km.head(), km.tail()
```

```python
plt.figure(figsize=(6,4))
plt.hist(km)
plt.title("Quilometragem dos corredores")
plt.xlabel("Quilometros")
plt.ylabel("Quantidade de corredores")
plt.show()

faixas = [0,5,12,21,42]
plt.figure(figsize=(6,4))
plt.hist(km,bins = faixas, edgecolor = "black")
plt.title("Quilometragem dos corredores")
plt.xlabel("Quilometros")
plt.ylabel("Quantidade de corredores")
plt.show()
```
